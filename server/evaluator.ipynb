```python
import numpy as np
import matplotlib.pyplot as plt


from time import sleep

from IPython.display import clear_output

from src.db_registry import registry
```

```python
def show_progress(last_indices):
    race_times = [race_data.get_race_time() for race_data in registry.get_race_data_objects()]
    
    if last_indices:
        race_times = race_times[len(race_times)-last_indices:]
    
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15, 10))
    ax.plot(race_times, color=(0.4, 0.0, 0.0))
    ax.set_title("Total race time")
    plt.show()
```

```python
update_interval = 30.0

while 1:
    clear_output(wait=True)
    show_progress(0)
    show_progress(1000)
    sleep(update_interval)
```
